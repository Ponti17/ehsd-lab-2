```python
resolution = 2*10**(-10);
fs = 1/resolution;
time_scale = 10**9;

pre_idx = 70;
post_idx = 220;
zoom_N = pre_idx + post_idx;
time_zoom = np.linspace(0, (zoom_N/fs)*time_scale, zoom_N);

color_b = [0, 0.4470, 0.7410];
color_r = [0.8500, 0.3250, 0.0980];
color_o = [0.9290, 0.6940, 0.1250];
```

```python
QNF_1_GND_trig_medi_idx = np.where(QNF_1_GND_trig >= np.median(QNF_1_GND_trig))[0][0]
QNF_1_GND_sig_zoom = QNF_1_GND_sig[QNF_1_GND_trig_medi_idx-pre_idx:QNF_1_GND_trig_medi_idx+post_idx]
QNF_1_VCC_trig_medi_idx = np.where(QNF_1_VCC_trig >= np.median(QNF_1_VCC_trig))[0][0]
QNF_1_VCC_sig_zoom = QNF_1_VCC_sig[QNF_1_VCC_trig_medi_idx-pre_idx:QNF_1_VCC_trig_medi_idx+post_idx]

SOIC_1_GND_trig_medi_idx = np.where(SOIC_1_GND_trig >= np.median(SOIC_1_GND_trig))[0][0]
SOIC_1_GND_sig_zoom = SOIC_1_GND_sig[SOIC_1_GND_trig_medi_idx-pre_idx:SOIC_1_GND_trig_medi_idx+post_idx]
SOIC_1_VCC_trig_medi_idx = np.where(SOIC_1_VCC_trig >= np.median(SOIC_1_VCC_trig))[0][0]
SOIC_1_VCC_sig_zoom = SOIC_1_VCC_sig[SOIC_1_VCC_trig_medi_idx-pre_idx:SOIC_1_VCC_trig_medi_idx+post_idx]
```

```python
fig1, ax = plt.subplots(2,2, sharex=True)
fig1.set_figheight(6)
fig1.set_figwidth(9)

ax[0,0].plot(time_zoom,QNF_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,0].plot(time_zoom,QNF_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);
ax[0,1].plot(time_zoom,SOIC_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,1].plot(time_zoom,SOIC_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);

# axis ticks
y_tic_gnd_7 = [-250, 0, 250, 500];
y_tic_vcc_7 = [2250, 2500, 2750, 3000, 3250];
ax.flat[0].set_yticks(y_tic_gnd_7)
ax.flat[1].set_yticks(y_tic_gnd_7)
ax.flat[2].set_yticks(y_tic_vcc_7)
ax.flat[3].set_yticks(y_tic_vcc_7)

ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

ax[0,0].set_title('QNF', fontsize=13)
ax[0,1].set_title('SOIC', fontsize=13)
ax[0,0].set_ylabel('GND', fontsize=13)
ax[1,0].set_ylabel('VCC', fontsize=13)

fig1.supylabel('Voltage [mV]')
fig1.supxlabel('Time [ns]')
fig1.suptitle('VCC/GND bounce 1 output', fontsize=16)
fig1.savefig('../latex/imgs/output_1.pdf', bbox_inches='tight')
```

```python
QNF_3_GND_trig_medi_idx = np.where(QNF_3_GND_trig >= np.median(QNF_3_GND_trig))[0][0]
QNF_3_GND_sig_zoom = QNF_3_GND_sig[QNF_3_GND_trig_medi_idx-pre_idx:QNF_3_GND_trig_medi_idx+post_idx]
QNF_3_VCC_trig_medi_idx = np.where(QNF_3_VCC_trig >= np.median(QNF_3_VCC_trig))[0][0]
QNF_3_VCC_sig_zoom = QNF_3_VCC_sig[QNF_3_VCC_trig_medi_idx-pre_idx:QNF_3_VCC_trig_medi_idx+post_idx]

SOIC_3_GND_trig_medi_idx = np.where(SOIC_3_GND_trig >= np.median(SOIC_3_GND_trig))[0][0]
SOIC_3_GND_sig_zoom = SOIC_3_GND_sig[SOIC_3_GND_trig_medi_idx-pre_idx:SOIC_3_GND_trig_medi_idx+post_idx]
SOIC_3_VCC_trig_medi_idx = np.where(SOIC_3_VCC_trig[130:-1] >= np.median(SOIC_3_VCC_trig))[0][0]
SOIC_3_VCC_sig_zoom = SOIC_3_VCC_sig[130:-1][SOIC_3_VCC_trig_medi_idx-pre_idx:SOIC_3_VCC_trig_medi_idx+post_idx]
```

```python
QNF_1_GND_sig_max = np.max(QNF_1_GND_sig)
print("QNF_1_GND_max: %0.3f " % (QNF_1_GND_sig_max*1000), "mV")

QNF_1_VCC_sig_min = np.min(QNF_1_VCC_sig)
print("QNF_1_VCC_min: %0.3f " % (QNF_1_VCC_sig_min*1000), "mV")

SOIC_1_GND_sig_max = np.max(SOIC_1_GND_sig)
print("SOIC_1_GND_max: %0.3f " % (SOIC_1_GND_sig_max*1000), "mV")

SOIC_1_VCC_sig_min = np.min(SOIC_1_VCC_sig)
print("SOIC_1_VCC_min: %0.3f " % (SOIC_1_VCC_sig_min*1000), "mV")

QNF_3_GND_sig_max = np.max(QNF_3_GND_sig)
print("QNF_3_GND_max: %0.3f " % (QNF_3_GND_sig_max*1000), "mV")

QNF_3_VCC_sig_min = np.min(QNF_3_VCC_sig)
print("QNF_3_VCC_min: %0.3f " % (QNF_3_VCC_sig_min*1000), "mV")

SOIC_3_GND_sig_max = np.max(SOIC_3_GND_sig)
print("SOIC_3_GND_max: %0.3f " % (SOIC_3_GND_sig_max*1000), "mV")

SOIC_3_VCC_sig_min = np.min(SOIC_3_VCC_sig)
print("SOIC_3_VCC_min: %0.3f " % (SOIC_3_VCC_sig_min*1000), "mV")

QNF_7_GND_sig_max = np.max(QNF_7_GND_sig)
print("QNF_7_GND_min: %0.3f " % (QNF_7_GND_sig_max*1000), "mV")

QNF_7_VCC_sig_min = np.min(QNF_7_VCC_sig)
print("QNF_7_VCC_min: %0.3f" % (QNF_7_VCC_sig_min*1000), "mV")

SOIC_7_GND_sig_max = np.max(SOIC_7_GND_sig)
print("SOIC_7_GND_max: %0.3f" % (SOIC_7_GND_sig_max*1000), "mV")

SOIC_7_VCC_sig_min = np.min(SOIC_7_VCC_sig)
print("SOIC_3_VCC_min: %0.3f" % (SOIC_7_VCC_sig_min*1000), "mV")

QNF_GND_sig_maxs = np.array([QNF_1_GND_sig_max, QNF_3_GND_sig_max, QNF_7_GND_sig_max])
QNF_VCC_sig_mins = np.array([QNF_1_VCC_sig_min, QNF_3_VCC_sig_min, QNF_7_VCC_sig_min])
SOIC_GND_sig_maxs = np.array([SOIC_1_GND_sig_max, SOIC_3_GND_sig_max, SOIC_7_GND_sig_max])
SOIC_VCC_sig_mins = np.array([SOIC_1_VCC_sig_min, SOIC_3_VCC_sig_min, SOIC_7_VCC_sig_min])

GND_sig_max_diffs = QNF_GND_sig_maxs - SOIC_GND_sig_maxs
GND_sig_max_diff_relative = (SOIC_GND_sig_maxs / QNF_GND_sig_maxs - 1) * 100
VCC_sig_min_diffs = QNF_VCC_sig_mins - SOIC_VCC_sig_mins
VCC_sig_min_diff_relative = ((3.3 - SOIC_VCC_sig_mins) /(3.3 - QNF_VCC_sig_mins) - 1) * 100

print()
print("Diffs on GND maxs, from QNF to SOIC: ", GND_sig_max_diffs)
print("Relative diffs on GND maxs, from SOCI to QNF: ", GND_sig_max_diff_relative)
print("Avg. Rel. diffs on GND maxs, from SOCI to QNF: ", np.mean(GND_sig_max_diff_relative))
print()
print("Diffs on VCC mins, from QNF to SOIC: ", VCC_sig_min_diffs)
print("Relative diffs on VCC mins, from SOCI to QNF: ", VCC_sig_min_diff_relative)
print("Avg. Rel. diffs on VCC mins, from SOCI to QNF: ", np.mean(VCC_sig_min_diff_relative))
```
