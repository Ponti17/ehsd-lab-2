```python
resolution = 2*10**(-10);
fs = 1/resolution;
time_scale = 10**9;

pre_idx = 70;
post_idx = 220;
zoom_N = pre_idx + post_idx;
time_zoom = np.linspace(0, (zoom_N/fs)*time_scale, zoom_N);

color_b = [0, 0.4470, 0.7410];
color_r = [0.8500, 0.3250, 0.0980];
color_o = [0.9290, 0.6940, 0.1250];
```

```python
QNF_1_GND_trig_medi_idx = np.where(QNF_1_GND_trig >= np.median(QNF_1_GND_trig))[0][0]
QNF_1_GND_sig_zoom = QNF_1_GND_sig[QNF_1_GND_trig_medi_idx-pre_idx:QNF_1_GND_trig_medi_idx+post_idx]
QNF_1_VCC_trig_medi_idx = np.where(QNF_1_VCC_trig >= np.median(QNF_1_VCC_trig))[0][0]
QNF_1_VCC_sig_zoom = QNF_1_VCC_sig[QNF_1_VCC_trig_medi_idx-pre_idx:QNF_1_VCC_trig_medi_idx+post_idx]

SOIC_1_GND_trig_medi_idx = np.where(SOIC_1_GND_trig >= np.median(SOIC_1_GND_trig))[0][0]
SOIC_1_GND_sig_zoom = SOIC_1_GND_sig[SOIC_1_GND_trig_medi_idx-pre_idx:SOIC_1_GND_trig_medi_idx+post_idx]
SOIC_1_VCC_trig_medi_idx = np.where(SOIC_1_VCC_trig >= np.median(SOIC_1_VCC_trig))[0][0]
SOIC_1_VCC_sig_zoom = SOIC_1_VCC_sig[SOIC_1_VCC_trig_medi_idx-pre_idx:SOIC_1_VCC_trig_medi_idx+post_idx]
```

```python
fig1, ax = plt.subplots(2,2, sharex=True)
fig1.set_figheight(6)
fig1.set_figwidth(9)

ax[0,0].plot(time_zoom,QNF_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,0].plot(time_zoom,QNF_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);
ax[0,1].plot(time_zoom,SOIC_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,1].plot(time_zoom,SOIC_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);

# axis ticks
y_tic_gnd_7 = [-250, 0, 250, 500];
y_tic_vcc_7 = [2250, 2500, 2750, 3000, 3250];
ax.flat[0].set_yticks(y_tic_gnd_7)
ax.flat[1].set_yticks(y_tic_gnd_7)
ax.flat[2].set_yticks(y_tic_vcc_7)
ax.flat[3].set_yticks(y_tic_vcc_7)

ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

ax[0,0].set_title('QNF', fontsize=13)
ax[0,1].set_title('SOIC', fontsize=13)
ax[0,0].set_ylabel('GND', fontsize=13)
ax[1,0].set_ylabel('VCC', fontsize=13)

fig1.supylabel('Voltage [mV]')
fig1.supxlabel('Time [ns]')
fig1.suptitle('VCC/GND bounce 1 output', fontsize=16)
fig1.savefig('../latex/imgs/output_1.pdf', bbox_inches='tight')
```

```python
import os

DATA_DIR = '../data/'
QFN_GND_1 = 'qfn_gnd_1.txt'
QFN_GND_3 = 'qfn_gnd_3.txt'
QFN_GND_7 = 'qfn_gnd_7.txt'
QFN_VCC_1 = 'qfn_vcc_1.txt'
QFN_VCC_3 = 'qfn_vcc_3.txt'
QFN_VCC_7 = 'qfn_vcc_7.txt'

QFN_1_GND_PATH = os.path.join(DATA_DIR, QFN_GND_1)
QFN_3_GND_PATH = os.path.join(DATA_DIR, QFN_GND_3)
QFN_7_GND_PATH = os.path.join(DATA_DIR, QFN_GND_7)
QFN_1_VCC_PATH = os.path.join(DATA_DIR, QFN_VCC_1)
QFN_3_VCC_PATH = os.path.join(DATA_DIR, QFN_VCC_3)
QFN_7_VCC_PATH = os.path.join(DATA_DIR, QFN_VCC_7)

df_qfn_gnd_1 = pd.read_csv(QFN_1_GND_PATH, sep='\t')
df_qfn_gnd_3 = pd.read_csv(QFN_3_GND_PATH, sep='\t')
df_qfn_gnd_7 = pd.read_csv(QFN_7_GND_PATH, sep='\t')
df_qfn_vcc_1 = pd.read_csv(QFN_1_VCC_PATH, sep='\t')
df_qfn_vcc_3 = pd.read_csv(QFN_3_VCC_PATH, sep='\t')
df_qfn_vcc_7 = pd.read_csv(QFN_7_VCC_PATH, sep='\t')

df_qfn_gnd = [df_qfn_gnd_1, df_qfn_gnd_3, df_qfn_gnd_7]
df_qfn_vcc = [df_qfn_vcc_1, df_qfn_vcc_3, df_qfn_vcc_7]
titles = ['QFN 1', 'QFN 3', 'QFN7']

fig, axs = plt.subplots(2, 3)
fig.set_figheight(7)
fig.set_figwidth(12)

axs[0,0].set_ylabel('VCC')
axs[1,0].set_ylabel('REF')
qfn_gnd_sig = [QNF_1_GND_sig_zoom, QNF_3_GND_sig_zoom, QNF_7_GND_sig_zoom]
qfn_vcc_sig = [QNF_1_VCC_sig_zoom, QNF_3_VCC_sig_zoom, QNF_7_VCC_sig_zoom]
shift_vcc = 10.5
shift_gnd = 11.2

for i, (df_gnd, df_vcc) in enumerate(zip(df_qfn_gnd, df_qfn_vcc)):
    axs[0, i].set_title(titles[i])
    axs[0, i].plot(df_vcc['time']*1e9, df_vcc['V(probe)'], linewidth=2, color='orange', alpha=0.5, label='Simulation')
    axs[0, i].plot(time_zoom-shift_vcc, qfn_vcc_sig[i], linewidth=2, color='blue', label='Measurement')
    axs[0, i].grid()
    axs[0, i].set_xlim(0, 20)
    axs[0, i].set_ylim(0.5, 4.5)
    axs[0, i].legend()

    axs[1, i].plot(df_gnd['time']*1e9, df_gnd['V(probe)'], linewidth=2, color='orange', alpha=0.5, label='Simulation')
    axs[1, i].plot(time_zoom-shift_gnd, qfn_gnd_sig[i], linewidth=2, color='blue', label='Measurement')
    axs[1, i].grid()
    axs[1, i].set_xlim(0, 20)
    axs[1, i].set_ylim(-1, 2)
    axs[1, i].legend()

fig.supylabel('Voltage [mV]')
fig.supxlabel('Time [ns]')

fig.tight_layout()
fig.savefig('../latex/imgs/sim_v_meas.pdf')
```
