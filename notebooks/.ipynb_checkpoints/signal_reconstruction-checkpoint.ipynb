```python
# load files and data

# QNF package, 1 output, GND bounce
QNF_1_GND = pd.read_csv('../lab_results/1_output_QNF_GND_bounceRefCurve_2024-12-03_1_095753.Wfm.csv', delimiter=';', names=['QNF_1_GND_sig', 'QNF_1_GND_trig'])
QNF_1_GND_sig = QNF_1_GND['QNF_1_GND_sig'].to_numpy();
QNF_1_GND_trig = QNF_1_GND['QNF_1_GND_trig'].to_numpy();

# QNF package, 1 output, VCC bounce
QNF_1_VCC = pd.read_csv('../lab_results/1_output_QNF_VCC_bounceRefCurve_2024-12-03_1_095502.Wfm.csv', delimiter=';', names=['QNF_1_VCC_sig', 'QNF_1_VCC_trig'])
QNF_1_VCC_sig = QNF_1_VCC['QNF_1_VCC_sig'].to_numpy();
QNF_1_VCC_trig = QNF_1_VCC['QNF_1_VCC_trig'].to_numpy();

# QNF package, 3 output, GND bounce
QNF_3_GND = pd.read_csv('../lab_results/3_output_QNF_GND_bounceRefCurve_2024-12-03_1_101256.Wfm.csv', delimiter=';', names=['QNF_3_GND_sig', 'QNF_3_GND_trig'])
QNF_3_GND_sig = QNF_3_GND['QNF_3_GND_sig'].to_numpy();
QNF_3_GND_trig = QNF_3_GND['QNF_3_GND_trig'].to_numpy();

# QNF package, 3 output, VCC bounce
QNF_3_VCC = pd.read_csv('../lab_results/3_output_QNF_VCC_BouhceRefCurve_2024-12-03_1_101516.Wfm.csv', delimiter=';', names=['QNF_3_VCC_sig', 'QNF_3_VCC_trig'])
QNF_3_VCC_sig = QNF_3_VCC['QNF_3_VCC_sig'].to_numpy();
QNF_3_VCC_trig = QNF_3_VCC['QNF_3_VCC_trig'].to_numpy();

# QNF package, 7 output, GND bounce
QNF_7_GND = pd.read_csv('../lab_results/7_output_QNF_GND_bounceRefCurve_2024-12-03_1_102240.Wfm.csv', delimiter=';', names=['QNF_7_GND_sig', 'QNF_7_GND_trig'])
QNF_7_GND_sig = QNF_7_GND['QNF_7_GND_sig'].to_numpy();
QNF_7_GND_trig = QNF_7_GND['QNF_7_GND_trig'].to_numpy();

# QNF package, 7 output, VCC bounce
QNF_7_VCC = pd.read_csv('../lab_results/7_output_QNF_VCC_bounceRefCurve_2024-12-03_1_102443.Wfm.csv', delimiter=';', names=['QNF_7_VCC_sig', 'QNF_7_VCC_trig'])
QNF_7_VCC_sig = QNF_7_VCC['QNF_7_VCC_sig'].to_numpy();
QNF_7_VCC_trig = QNF_7_VCC['QNF_7_VCC_trig'].to_numpy();


# SOIC package, 1 output, GND bounce
SOIC_1_GND = pd.read_csv('../lab_results/1_output_SOIC_GND_bounceRefCurve_2024-12-03_1_095943.Wfm.csv', delimiter=';', names=['SOIC_1_GND_sig', 'SOIC_1_GND_trig'])
SOIC_1_GND_sig = SOIC_1_GND['SOIC_1_GND_sig'].to_numpy();
SOIC_1_GND_trig = SOIC_1_GND['SOIC_1_GND_trig'].to_numpy();

# SOIC package, 1 output, VCC bounce
SOIC_1_VCC = pd.read_csv('../lab_results/1_output_SOIC_VCC_bounceRefCurve_2024-12-03_1_100139.Wfm.csv', delimiter=';', names=['SOIC_1_VCC_sig', 'SOIC_1_VCC_trig'])
SOIC_1_VCC_sig = SOIC_1_VCC['SOIC_1_VCC_sig'].to_numpy();
SOIC_1_VCC_trig = SOIC_1_VCC['SOIC_1_VCC_trig'].to_numpy();

# QNF package, 3 output, GND bounce
SOIC_3_GND = pd.read_csv('../lab_results/3_output_SOIC_GND_bounceRefCurve_2024-12-03_1_101745.Wfm.csv', delimiter=';', names=['SOIC_3_GND_sig', 'SOIC_3_GND_trig'])
SOIC_3_GND_sig = SOIC_3_GND['SOIC_3_GND_sig'].to_numpy();
SOIC_3_GND_trig = SOIC_3_GND['SOIC_3_GND_trig'].to_numpy();

# QNF package, 3 output, VCC bounce
SOIC_3_VCC = pd.read_csv('../lab_results/3_output_SOIC_VCC_bounceRefCurve_2024-12-03_1_102012.Wfm.csv', delimiter=';', names=['SOIC_3_VCC_sig', 'SOIC_3_VCC_trig'])
SOIC_3_VCC_sig = SOIC_3_VCC['SOIC_3_VCC_sig'].to_numpy();
SOIC_3_VCC_trig = SOIC_3_VCC['SOIC_3_VCC_trig'].to_numpy();

# QNF package, 7 output, GND bounce
SOIC_7_GND = pd.read_csv('../lab_results/7_output_SOIC_GND_bounceRefCurve_2024-12-03_1_102644.Wfm.csv', delimiter=';', names=['SOIC_7_GND_sig', 'SOIC_7_GND_trig'])
SOIC_7_GND_sig = SOIC_7_GND['SOIC_7_GND_sig'].to_numpy();
SOIC_7_GND_trig = SOIC_7_GND['SOIC_7_GND_trig'].to_numpy();

# QNF package, 7 output, VCC bounce
SOIC_7_VCC = pd.read_csv('../lab_results/7_output_SOIC_VCC_bounceRefCurve_2024-12-03_1_102905.Wfm.csv', delimiter=';', names=['SOIC_7_VCC_sig', 'SOIC_7_VCC_trig'])
SOIC_7_VCC_sig = SOIC_7_VCC['SOIC_7_VCC_sig'].to_numpy();
SOIC_7_VCC_trig = SOIC_7_VCC['SOIC_7_VCC_trig'].to_numpy();


# HEX Inverter self-oscillating 
INV_OSCI_1 = pd.read_csv('../lab_results/hex_invt_osci_1_outputRefCurve_2024-12-03_1_103825.Wfm.csv', delimiter=';', names=['INV_OSCI_1_sig', 'INV_OSCI_1_trig'])
INV_OSCI_1_sig = INV_OSCI_1['INV_OSCI_1_sig'].to_numpy();
INV_OSCI_1_trig = INV_OSCI_1['INV_OSCI_1_trig'].to_numpy();

INV_OSCI_7 = pd.read_csv('../lab_results/hex_invt_osci_7_outputRefCurve_2024-12-03_1_103825.Wfm.csv', delimiter=';', names=['INV_OSCI_7_sig', 'INV_OSCI_7_trig'])
INV_OSCI_7_sig = INV_OSCI_7['INV_OSCI_7_sig'].to_numpy();
INV_OSCI_7_trig = INV_OSCI_7['INV_OSCI_7_trig'].to_numpy();
```

```python
resolution = 2*10**(-10);
fs = 1/resolution;
time_scale = 10**9;

pre_idx = 70;
post_idx = 220;
zoom_N = pre_idx + post_idx;
time_zoom = np.linspace(0, (zoom_N/fs)*time_scale, zoom_N);

color_b = [0, 0.4470, 0.7410];
color_r = [0.8500, 0.3250, 0.0980];
color_o = [0.9290, 0.6940, 0.1250];
```

```python
QNF_1_GND_trig_medi_idx = np.where(QNF_1_GND_trig >= np.median(QNF_1_GND_trig))[0][0]
QNF_1_GND_sig_zoom = QNF_1_GND_sig[QNF_1_GND_trig_medi_idx-pre_idx:QNF_1_GND_trig_medi_idx+post_idx]
QNF_1_VCC_trig_medi_idx = np.where(QNF_1_VCC_trig >= np.median(QNF_1_VCC_trig))[0][0]
QNF_1_VCC_sig_zoom = QNF_1_VCC_sig[QNF_1_VCC_trig_medi_idx-pre_idx:QNF_1_VCC_trig_medi_idx+post_idx]

SOIC_1_GND_trig_medi_idx = np.where(SOIC_1_GND_trig >= np.median(SOIC_1_GND_trig))[0][0]
SOIC_1_GND_sig_zoom = SOIC_1_GND_sig[SOIC_1_GND_trig_medi_idx-pre_idx:SOIC_1_GND_trig_medi_idx+post_idx]
SOIC_1_VCC_trig_medi_idx = np.where(SOIC_1_VCC_trig >= np.median(SOIC_1_VCC_trig))[0][0]
SOIC_1_VCC_sig_zoom = SOIC_1_VCC_sig[SOIC_1_VCC_trig_medi_idx-pre_idx:SOIC_1_VCC_trig_medi_idx+post_idx]
```

```python
fig1, ax = plt.subplots(2,2, sharex=True)
fig1.set_figheight(6)
fig1.set_figwidth(9)

ax[0,0].plot(time_zoom,QNF_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,0].plot(time_zoom,QNF_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);
ax[0,1].plot(time_zoom,SOIC_1_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,1].plot(time_zoom,SOIC_1_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);

# axis ticks
y_tic_gnd_7 = [-250, 0, 250, 500];
y_tic_vcc_7 = [2250, 2500, 2750, 3000, 3250];
ax.flat[0].set_yticks(y_tic_gnd_7)
ax.flat[1].set_yticks(y_tic_gnd_7)
ax.flat[2].set_yticks(y_tic_vcc_7)
ax.flat[3].set_yticks(y_tic_vcc_7)

ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

ax[0,0].set_title('QNF', fontsize=13)
ax[0,1].set_title('SOIC', fontsize=13)
ax[0,0].set_ylabel('GND', fontsize=13)
ax[1,0].set_ylabel('VCC', fontsize=13)

fig1.supylabel('Voltage [mV]')
fig1.supxlabel('Time [ns]')
fig1.suptitle('VCC/GND bounce 1 output', fontsize=16)
fig1.savefig('../latex/imgs/output_1.pdf', bbox_inches='tight')
```

```python
QNF_1_GND_sig_no_mean = QNF_1_GND_sig - np.mean(QNF_1_GND_sig) 
QNF_1_GND_sig_min = np.min(QNF_1_GND_sig_no_mean)
QNF_1_GND_sig_max = np.max(QNF_1_GND_sig_no_mean)
print("QNF_1_GND_min: ", QNF_1_GND_sig_min)
print("QNF_1_GND_max: ", QNF_1_GND_sig_max)
print()

QNF_1_VCC_sig_no_mean = QNF_1_VCC_sig - np.mean(QNF_1_VCC_sig) 
QNF_1_VCC_sig_min = np.min(QNF_1_VCC_sig_no_mean)
QNF_1_VCC_sig_max = np.max(QNF_1_VCC_sig_no_mean)
print("QNF_1_VCC_min: ", QNF_1_VCC_sig_min)
print("QNF_1_VCC_max: ", QNF_1_VCC_sig_max)
print()

SOIC_1_GND_sig_no_mean = SOIC_1_GND_sig - np.mean(QNF_1_GND_sig) 
SOIC_1_GND_sig_min = np.min(SOIC_1_GND_sig_no_mean)
SOIC_1_GND_sig_max = np.max(SOIC_1_GND_sig_no_mean)
print("SOIC_1_GND_min: ", SOIC_1_GND_sig_min)
print("SOIC_1_GND_max: ", SOIC_1_GND_sig_max)
print()

SOIC_1_VCC_sig_no_mean = SOIC_1_VCC_sig - np.mean(QNF_1_VCC_sig) 
SOIC_1_VCC_sig_min = np.min(SOIC_1_VCC_sig_no_mean)
SOIC_1_VCC_sig_max = np.max(SOIC_1_VCC_sig_no_mean)
print("SOIC_1_VCC_min: ", SOIC_1_VCC_sig_min)
print("SOIC_1_VCC_max: ", SOIC_1_VCC_sig_max)
print()
```

```python
QNF_3_GND_trig_medi_idx = np.where(QNF_3_GND_trig >= np.median(QNF_3_GND_trig))[0][0]
QNF_3_GND_sig_zoom = QNF_3_GND_sig[QNF_3_GND_trig_medi_idx-pre_idx:QNF_3_GND_trig_medi_idx+post_idx]
QNF_3_VCC_trig_medi_idx = np.where(QNF_3_VCC_trig >= np.median(QNF_3_VCC_trig))[0][0]
QNF_3_VCC_sig_zoom = QNF_3_VCC_sig[QNF_3_VCC_trig_medi_idx-pre_idx:QNF_3_VCC_trig_medi_idx+post_idx]

SOIC_3_GND_trig_medi_idx = np.where(SOIC_3_GND_trig >= np.median(SOIC_3_GND_trig))[0][0]
SOIC_3_GND_sig_zoom = SOIC_3_GND_sig[SOIC_3_GND_trig_medi_idx-pre_idx:SOIC_3_GND_trig_medi_idx+post_idx]
SOIC_3_VCC_trig_medi_idx = np.where(SOIC_3_VCC_trig[130:-1] >= np.median(SOIC_3_VCC_trig))[0][0]
SOIC_3_VCC_sig_zoom = SOIC_3_VCC_sig[130:-1][SOIC_3_VCC_trig_medi_idx-pre_idx:SOIC_3_VCC_trig_medi_idx+post_idx]
```

```python
fig3, ax = plt.subplots(2,2, sharex=True)
fig3.set_figheight(6)
fig3.set_figwidth(9)

ax[0,0].plot(time_zoom,QNF_3_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,0].plot(time_zoom,QNF_3_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);
ax[0,1].plot(time_zoom,SOIC_3_GND_sig_zoom*1000, color=color_b, linewidth=1.2);
ax[1,1].plot(time_zoom,SOIC_3_VCC_sig_zoom*1000, color=color_r, linewidth=1.2);

# axis ticks
y_tic_gnd_7 = [-500, -250, 0, 250, 500, 750, 1000];
y_tic_vcc_7 = [2000, 2250, 2500, 2750, 3000, 3250, 3500];
ax.flat[0].set_yticks(y_tic_gnd_7)
ax.flat[1].set_yticks(y_tic_gnd_7)
ax.flat[2].set_yticks(y_tic_vcc_7)
ax.flat[3].set_yticks(y_tic_vcc_7)

ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

ax[0,0].set_title('QNF', fontsize=13)
ax[0,1].set_title('SOIC', fontsize=13)
ax[0,0].set_ylabel('GND', fontsize=13)
ax[1,0].set_ylabel('VCC', fontsize=13)

fig3.supylabel('Voltage [mV]')
fig3.supxlabel('Time [ns]')
fig3.suptitle('VCC/GND bounce 3 output', fontsize=16)
fig3.savefig('../latex/imgs/output_3.pdf', bbox_inches='tight')
```
